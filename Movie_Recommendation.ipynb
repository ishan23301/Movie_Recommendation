```python
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
        
```

```python
movie.genres = movie.genres.apply(convert)
```

```python
def director(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
        else:
            continue
    return list
```

```python
movie.crew = movie.crew.apply(director)
```

```python
def cast(obj):
    list = []
    count = 0
    for i in ast.literal_eval(obj):
        if(count<3):
            list.append(i['name'])
            count += 1
            

    return list
            
```

```python
movie.cast = movie.cast.apply(cast)
```

```python
movie['overview'] = movie['overview'].apply(lambda x: x.split(' '))
```

```python
movie.head()
```

```python
movie['genres'] = movie['genres'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movie['keywords'] = movie['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movie['cast'] = movie['cast'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movie['crew'] = movie['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movie.head()
```

```python
movie['tag'] = movie['overview'] + movie['keywords'] + movie['cast'] + movie['crew'] + movie['genres']
```

```python
new_movie = movie[['movie_id', 'title', 'tag']] 
```

```python
new_movie
```

```python
new_movie['tag'] = new_movie['tag'].apply(lambda x:" ".join(x))
```

```python
new_movie['tag'][0]
```

```python
new_movie['tag'] = new_movie['tag'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split(" "):
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movie['tag'] = new_movie['tag'].apply(stem)
```

```python
new_movie.tag[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_movie['tag']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_movie[new_movie['title']==movie].index[0]
    sim = similarity[movie_index]
    movies_list = sorted(list(enumerate(sim)), reverse =True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_movie.iloc[i[0]].title)
    
    
```

```python
recommend('Krrish')
```
